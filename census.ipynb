```python
for i in df.columns:
    if df[i].dtypes == 'object':
        le2 = LabelEncoder()
        df[i] = le2.fit_transform(df[i])
```

```python
df
```

```python
#Feature Selection
```

```python
x = df.drop('Income', axis=1)
y = df['Income']
```

```python
x.info()
```

```python
x = x.drop(['Workclass', 'Education', 'Race', 'Sex','Capital_loss','Native_country'], axis=1)
```

```python
x
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
for col in x.columns:
    scaler = StandardScaler()
    x[col] = scaler.fit_transform(x[col].values.reshape(-1, 1))
```

```python
 #Fixing imbalanced dataset using Oversampling
```

```python
y
```

```python
round(y.value_counts(normalize=True) * 100, 2).astype('str') + ' %'
```

```python
from imblearn.over_sampling import RandomOverSampler
ro = RandomOverSampler(random_state=42)
```

```python
ro.fit(x, y)
```

```python
x_new, y_new = ro.fit_resample(x, y)
```

```python
round(y_new.value_counts(normalize=True) * 100, 2).astype('str') + ' %'
```

```python
#Creating a train test split
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    x_new, y_new, test_size=0.3, random_state=42)
```

```python
print("x_train shape:", x_train.shape)
print("x_test shape:", x_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
#Data Modelling and evaluation
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
```

```python
 #Logistic Regression
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression(random_state=42)
```

```python
LR.fit(x_train, y_train)
```

```python
y_pred_LR = LR.predict(x_test)
```

```python
print('Accuracy score:', round(accuracy_score(y_test, y_pred_LR) * 100, 2))
print('F1 score:', round(f1_score(y_test, y_pred_LR) * 100, 2))
```

```python
#Hyperparameter Tuning as the Random forest classifier is havig the heigher accuracy
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
n_estimators = [int(x) for x in np.linspace(start=40, stop=150, num=15)]
max_depth = [int(x) for x in np.linspace(40, 150, num=15)]
```

```python
param_dist = {'n_estimators': n_estimators,'max_depth': max_depth,}
```

```python
RC_tuned = RandomForestClassifier(random_state=42)
```

```python
RC_cv = RandomizedSearchCV(
    estimator=RC_tuned, param_distributions=param_dist, cv=5, random_state=42)
```

```python
RC_cv.fit(x_train, y_train)
```

```python
RC_cv.best_score_
```

```python
RC_cv.best_params_
```

```python
RC_best = RandomForestClassifier(
    max_depth=102, n_estimators=40, random_state=42)
```

```python
RC_best.fit(x_train, y_train)
```

```python
Y_pred_RC_BEST = RC_best.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, Y_pred_RC_BEST)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, Y_pred_RC_BEST))
```
